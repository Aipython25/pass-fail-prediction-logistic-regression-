```python

from google.colab import files
uploaded = files.upload()
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
data = pd.read_csv("salary_data.csv")
print("Dataset:")
print(data.head())
X = data[['Experience', 'Test_Score', 'Interview_Score']]
y = data['Salary']
model = LinearRegression()
model.fit(X, y)
input_data = [[ 5, 88, 75]]
predicted_salary = model.predict(input_data)
print("Predicted salary for 5 exp, 88 test score, 75 interview score:" , round(predicted_salary[0], 2))
joblib.dump(model, 'salary_predictor_model.pkl')
model =  joblib.load('salary_predictor_model.pkl')
```
